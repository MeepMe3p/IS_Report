```python
import pandas as pd

# Replace this with the path to your CSV file
df = pd.read_csv("../data/csv/filtered_dataset.csv")

# Assuming the first two columns are non-labels (like 'id' and something else)
label_columns = df.columns[2:]

# Find rows where more than one label is 1
multi_label_rows = df[df[label_columns].sum(axis=1) > 1]

# Display the first such row
print(multi_label_rows.head(1))
```

```python


df = pd.read_csv("../data/csv/filtered_dataset.csv")

df_melt = df.melt(id_vars=['subject_id'], var_name="Feature", value_name="Label")

df_ones = df_melt[df_melt['Label'] == 1]

# Plot
plt.figure(figsize=(20, 10))
sns.countplot(data=df_ones, x="Feature", order=df_ones["Feature"].value_counts().index)
plt.title("Count of 1s for each feature")
plt.show()
```
